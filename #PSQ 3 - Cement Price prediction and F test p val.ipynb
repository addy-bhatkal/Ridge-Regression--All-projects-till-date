```python
x = df[['cement', 'slag', 'flyash', 'water', 'superplasticizer','age']]
x = sm.add_constant(x)
y = df.iloc[:, -1] 
```

```python
model4 = sm.OLS(y, x).fit()
print(model4.summary())
```

```python
lr = LinearRegression()
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
model2 = lr.fit(x_train, y_train)
y_predict = lr.predict(x_test)
r2 = r2_score(y_predict, y_test)
r2
ssm = np.sum((y_test.mean()- y_test)**2) 
df1 = x_train.shape[1] - 1 
ssf = np.sum((y_predict - y_test)**2)
df2 = x_train.shape[0] - x_train.shape[1]
fstat = ((ssm-ssf)/df1) / (ssf/df2) 
pval = stats.f.sf(fstat, df1, df2)
pval
```

```python
#cross val n Linear Reg random state values 
for i in range(1,1000):
    x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=i)
    cvv = cross_val_score(lr, x_train, y_train, cv = 5, scoring = 'r2')
    if cvv.mean() > 0.63:
        print(i, cvv.mean()) 
```

```python
#Best alpha score using cross val on training data
alphas = np.logspace(-6,6,10)
for index,i in enumerate(alphas):
    rd = Ridge(alpha=i) 
    cvv2 = cross_val_score(rd, x_train, y_train, cv = 5, scoring = 'r2')
    print(i, cvv2.mean())
```

```python
#Using for loop on random state values in Ridge, with alpha = -101 
for i in range(1,10000):
    x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=i)
    rd = Ridge(alpha=-101)
    model3 = rd.fit(x_train, y_train)
    y_predict = model3.predict(x_test)
    r2 = r2_score(y_predict, y_test)
    if r2>0.63:
        print(i, r2)
```
